### Scalar

```python
import numpy as np 
x = np.array(12)
print(x.ndim)
print(x)
```

### Vector

```python
x = np.array([12,3,6,14])
print(x)
print(x.ndim)
```

### Matrices

```python
x = np.array([[5,78,2,34,0],[6,79,3,25,1],[1,2,3,4,5]])
print(x.ndim)
print(x)
```

### Shapes

```python
from keras.datasets import mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
print(train_images.ndim)
print(train_images.shape)
print(train_images.dtype)
```

```python
digit = train_images[4]

import matplotlib.pyplot as plt 
plt.imshow(digit, cmap=plt.cm.binary)
plt.show()
```

```python
my_slice = train_images[10:100]
print(my_slice.shape)
```

```python
digit_cut = train_images[4, 14:, 14:]
plt.imshow(digit_cut, cmap=plt.cm.binary)
plt.show()
digit_cut = train_images[4, 7:-7, 7:-7]
plt.imshow(digit_cut, cmap=plt.cm.binary)
plt.show()



```
